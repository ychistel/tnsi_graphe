```python
from pile import Pile,creer_pile

def parcours_profondeur(g,s):
    vus=[]
    p=creer_pile()
    p.empiler(s)
    while not p.est_vide():
        print(p,vus)
        s=p.depiler()
        if s in vus:
            continue
        vus.append(s)
        for v in g.voisins(s):
            p.empiler(v)
    return vus

def chemin(g,s,t):
    parcours=parcours_profondeur(g,s)
    if t in parcours:
        return parcours[parcours.index(s):parcours.index(t)+1]
    
H={'A':{'B','C'},\
   'B':{'D','E'},\
   'C':{'F','E'},\
   'D':{'B','E','G'},\
   'E':{'B','D','C','G'},\
   'F':{'C','H'},\
   'G':{'D','E','H'},\
   'H':{'F','G'}
  }
H=dict_to_graphe(H)
print(H.dict.valeur)


parcours_profondeur(H,'F')
H.afficher()
```

```python
def parcours_largeur(g,source):
    dist={source:0}
    courant={source}
    suivant=set()
    while len(courant)>0:
        s=courant.pop()
        for v in g.voisins(s):
            if v not in dist:
                suivant.add(v)
                dist[v]=dist[s]+1
            if len(courant)==0:
                courant,suivant=suivant,set()
    return dist

def distance(g,s1,s2):
    dist=parcours_largeur(g,s1)
    return dist[s2] if s2 in dist else None
```

```python
def parcours_largeur(g,s):
    dist={s:0}
    courant={s}
    suivant=set()
    while len(courant)>0:
        sommet=courant.pop()
        for v in g.voisins(sommet):
            if v not in dist:
                suivant.add(v)
                dist[v]=dist[sommet]+1
            if len(courant)==0:
                courant,suivant=suivant,set()
    return dist
```
