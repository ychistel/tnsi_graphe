````python
"""------- Modules -------"""

from random import randint

"""------- Variables-----------"""

# nombre total d'allumettes
n = 14

# plateau de jeu
nim = [1 for _ in range(n)]

# nombre d'allumettes à retirer
nb_al = 1 # max 3
position_al = 0

# joueur ou ordinateur qui joue
joueur = True # le joueur joue, si False, l'ordinateur joue
````

````python
# Fonctions du jeu

def saisir_nb_al(joueur,reste_al):
    nb_al=0
    if joueur:
        while nb_al <= 0 or nb_al > 3 or nb_al >= reste_al :
            nb_al = int(input("Combien d'allumettes voulez-vous enlever ? "))
    else:
        while nb_al <= 0 or nb_al > 3 or nb_al >= reste_al :
            nb_al = randint(1,3)
    return nb_al
````

````python
"""-------- Boucle de jeu --------"""

# tant qu'il y a des allumettes on joue
while sum(nim)>=1:
    debut = position_al
    reste_al = sum(nim)
    if reste_al == 1:
        if joueur:
            print("Vous avez perdu, la machine gagne")
        else:
            print("Vous avez gagné, la machine perd")
        break
    else:
        if joueur:
            nb_al = saisir_nb_al(joueur,reste_al)
            fin = position_al + nb_al
            if nb_al < reste_al:
                for i in range(debut,fin):
                    nim[i] = 0
            joueur=False
        else:
            m = min(reste_al,3)
            nb_al = saisir_nb_al(joueur,reste_al)
            fin = position_al + nb_al
            if nb_al < reste_al:
                for i in range(debut,fin):
                    nim[i] = 0
            joueur=True
        
        position_al += nb_al
        print(debut,fin,nim)
````

## Jeu de NIM avec un graphe

On a vu qu'un jeu de NIM peut se représenter par un graphe dont chaque sommet correspond au nombre d'allumettes restantes. Les sommets adjacents sont les nombres d'allumettes obtenus en enlevant 1, 2 ou 3 allumettes.

On définit le dictionnaire d'adjacence du graphe par le dictionnaire suivant:

```python

# On construit le graphe d'adjacence du jeu

    jeu_nim[v] = {v-1,v-2,v-3}
jeu_nim[3]={1,2}
jeu_nim[2]={1}
jeu_nim[1]={0}
```

````python
from random import randint

# Fonctions du jeu
def construire_graphe(n):
    """
    Le graphe du jeu de NIM est un dictionnaire d'adjacence donnant les différentes étapes du jeu.
    """
    graphe = {}
    graphe[1]={0}
    graphe[2]={1}
    graphe[3]={1,2}
    for i in range(4,n+1):
        graphe[i] = {i-1,i-2,i-3}
    return graphe
    
def noyau_du_graphe(n):
    noyau = [1]
    while noyau[-1] < n:
        noyau.append(noyau[-1]+4)
    return noyau

def saisir_nb_al(joueur,jeu,k):
    nb_al=0
    if joueur:
        while k-nb_al not in jeu :
            nb_al = int(input("Combien d'allumettes voulez-vous enlever ? "))
    else:
        while k-nb_al not in jeu :
            nb_al = randint(1,3)
    return nb_al

def afficher_allumettes(k):
    tete = "O  "*k
    baton = "|  "*k
    dessin = tete + "\n" + baton + "\n" + baton +"\n" + baton + "\n"
    return dessin



"""-------- Variables ------------"""
# Nombre d'allumettes au début du jeu
n=20


"""-------- Boucle de jeu --------"""

# On crée le graphe de la partie du jeu de NIM pour n allumettes
jeu_nim = construire_graphe(n)


# On détermine le noyau du graphe (nombre d'allumettes à atteindre pour gagner)
noyau = noyau_du_graphe(n)

# La variable jeu contient les états du jau au coup prochain à jouer
jeu = jeu_nim[n]

# La variable joueur est un booléen. Si vrai, le joueur joue, si false, la machine joue
# Ici, le joueur commence
joueur = True

# k est la variable indiquant le nombre d'allumettes restant
k=n


# tant qu'il y a des allumettes on joue
while k>0:
    """
    S'il ne reste qu'une seule allumette, la partie est finie:
    - si joueur = True, c'est au joueur de jouer donc il a perdu !
    - si joueur = False, c'est à la machine de jouer, alle a perdu!
    
    La machine connaît la stratégie gagnante. Elle retire le nombre d'allumettes
    pour être dans le noyau du graphe ! Si ce n'est pas possible, elle retire un
    nombre aléatoire d'allumettes
    """
   
    if k == 1:
        if joueur:
            print("Il reste 1 allumette, vous avez perdu, la machine gagne.")
        else:
            print("Il reste 1 allumette, vous avez gagné, la machine a perdu.")
        break
        
    # Il reste plus qu'une allumette, la partie continue
    else:
        if k == n:
            print(afficher_allumettes(k))
            print("Le jeu commence. Il y a %s allumettes sur le tapis.\nC'est au joueur de commencer !" % n)
        # C'est au joueur de jouer:
        if joueur:
            nb_al = saisir_nb_al(joueur,jeu,k) # le joueur choisit le nombre d'allumettes qu'il veut enlever
            k = k - nb_al # on enlève les allumettes, il en reste k
            jeu=jeu_nim[k] # on actualise l'état du jeu avec le graphe
            print("Vous enlevez %s allumettes. Il en reste %s." % (nb_al,k))
            joueur=False # au prochain tour, c'est à la machine de joueur
            
        # C'est à la machine de jouer
        else:            
            # La machine doit enlever le nombre d'allumettes pour être dans le noyau du graphe:
            if k in noyau: # k est déjà dans le noyau, la machine ne peut pas atteindre une autre valeur du noyau !
                nb_al = saisir_nb_al(joueur,jeu,k) # elle choisit le nombre d'allumettes à retirer au hasard
                k = k - nb_al # on enlève les allumettes, il en reste k
            else:
                # la machine peut atteindre le noyau
                i = 1
                while noyau[i] < k:
                    i += 1
                nb_al = k - noyau[i-1]
                k = noyau[i-1]
            jeu = jeu_nim[k] # on actualise l'état du jeu avec le graphe
            print("La machine enlève %s allumettes. Il en reste %s." % (nb_al,k))
            joueur=True # au prochain tour, c'est au joueur de joueur

        print(afficher_allumettes(k))
        
````
